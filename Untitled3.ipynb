```python
import pandas as pd
import seaborn as sns

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.read_csv('dataset_siswa.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
sns.heatmap(df.isna(), cbar=False, cmap='viridis')
plt.title("Heatmap Missing Values")
plt.show()
```

```python
df['Nilai_Matematika'] = df['Nilai_Matematika'].fillna(df['Nilai_Matematika'].mean())
df['Nilai_Bahasa_Indonesia'] = df['Nilai_Bahasa_Indonesia'].fillna(df['Nilai_Bahasa_Indonesia'].mean())
df['Nilai_Informatika'] = df['Nilai_Informatika'].fillna(df['Nilai_Informatika'].mean())
df['Rata_Rata'] = df['Rata_Rata'].fillna(df['Rata_Rata'].mean())

df['Jurusan'] = df['Jurusan'].fillna(df['Jurusan'].mode()[0])
df['Kelas'] = df['Kelas'].fillna(df['Kelas'].mode()[0])
df['Nama'] = df['Nama'].fillna(df['Nama'].mode()[0])
df['NIS'] = df['NIS'].fillna(df['NIS'].mode()[0])

```

```python
df = df.drop(columns=['deck'], errors='ignore')

```

```python
df.to_csv('titanic_bersih.csv', index=False)
```
